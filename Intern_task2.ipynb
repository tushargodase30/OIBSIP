Unemloyment Analysis With Python

Unemployment is measured by the unemployment rate which is the number of people who are unemployed as a percentage of the total labour force. We have seen a sharp increase in the unemployment rate during Covid-19, so analyzing the unemployment rate can be a good data science project.

#Data Visualization

```python
import plotly.express as px
import plotly
```

```python
values = data[' Estimated Unemployment Rate (%)']
category = data['Region']
```

#Pie Charts

```python
fig = px.pie(data , values = values , names=category , title='Estimated Unemployment Rate (%)')
fig.update_traces(textposition="inside" , textinfo="percent")
fig.update_layout(autosize=False,
    width = 800,
    height = 800,
    title_font_size = 20
)
```

```python
values=data[' Estimated Employed']
category=data['Region']
```

```python
fig = px.pie(data , values = values , names=category , title='Estimated Employed')
fig.update_traces(textposition="inside" , textinfo="percent")
fig.update_layout(autosize=False,
    width = 800,
    height = 800,
    title_font_size = 20
)
```

```python
values = data[' Estimated Labour Participation Rate (%)']
category = data['Region']

fig = px.pie(data , values=values , names=category  , title='Estimated Labour Participation Rate (%)')
fig.update_layout(
    autosize = False,
    width = 800,
    height = 800
)
```

#Barplots

```python
import warnings
warnings.filterwarnings('ignore')
fig , aixs = plt.subplots(figsize = (10, 10))
plt.subplot(2 ,2  ,1)
sns.barplot(x ='Region.1' , y =' Estimated Unemployment Rate (%)' , data = data)
plt.subplot(2 , 2 , 2)
sns.barplot(x ='Region.1' , y =' Estimated Employed' , data = data)
plt.subplot(2 , 2 , 3)
sns.barplot(x ='Region.1' , y=' Estimated Labour Participation Rate (%)' , data = data)
plt.tight_layout()
```

#Boxplots

```python
fig , axis = plt.subplots(figsize=(8 , 8))
plt.subplot(2 ,2 ,1)
sns.boxplot(x ='Region.1' , y =' Estimated Unemployment Rate (%)' , data = data )
plt.subplot(2 ,2 , 2)
sns.boxplot(x ='Region.1' , y =' Estimated Employed' , data = data )
plt.subplot(2 , 2 , 3)
sns.boxplot(x ='Region.1' , y =' Estimated Labour Participation Rate (%)' , data = data)
plt.tight_layout()
```
